###2. Import Library Twint

```python
import twint
import pandas
import nest_asyncio
nest_asyncio.apply()
```

###3. Crawling Data

```python
c = twint.Config()
c.Search = "Cak Imin"
c.Since = "2023-01-01"
c.Until = "2023-02-16"
c.Limit = 500
c.Pandas = True
c.Lang = "id"

twint.run.Search(c)
```

###4. Data Frame

```python
data = twint.storage.panda.Tweets_df
df = pandas.DataFrame(data, columns=['date', 'username', 'tweet'])
df
```
